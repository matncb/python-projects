```python
# filtros

def aplicar_filtro(df):
    filtro = df

    filtro = filtro[filtro['P/L'] > 3]
    filtro = filtro[filtro['P/L'] < 15]

    filtro = filtro[filtro['P/VP'] < 2]
    filtro = filtro[filtro['P/VP'] > 0.5]

    filtro = filtro[filtro['Div.Yield'] > 9/100]
    filtro = filtro[filtro['Div.Yield'] < 30/100]

    filtro = filtro[filtro['ROE'] > 10/100]

    filtro = filtro[filtro['Liq.2meses'] > 1e6]

    filtro = filtro[filtro['Cresc. Rec.5a'] > 10/100]

    filtro = filtro.sort_values(["P/L", "P/VP", "Div.Yield", 'ROE', 'Dív.Brut/ Patrim.' ], ascending=[True, True, False, False, True])

    return filtro

filtro = aplicar_filtro(df)
filtro
```
